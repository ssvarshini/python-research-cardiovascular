# 2 Web Scrapping

## Extracting HTML Content:
Firstly, I check if the webpage content loads correctly and ensure access is not denied. This involves fetching the HTML content of the webpage containing the required data using the requests library. Once the content is retrieved, I use BeautifulSoup to parse the HTML structure, making it ready for extracting the relevant table data for analysis. I will print out the first 20 lines to check if the website is loaded.

```python
url = 'https://bmcmedinformdecismak.biomedcentral.com/articles/10.1186/s12911-020-1023-5/tables/1'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

#Print the first few lines of the parsed HTML
html_lines = soup.prettify().splitlines()
for line in html_lines[:20]:  
    print(line)
```

## Extracting Tabular Data

While the webpage content has been successfully loaded, the raw output contains the entire page's information, much of which is redundant for the analysis. My focus is on extracting specific data within a table. Using the parsed HTML structure, I locate the relevant table that contains the data of interest. This step ensures that only the necessary information is isolated for further processing, eliminating noise and redundancy from the rest of the webpage content.

```python
table = soup.find_all('table')[0]
print(table)
```

Now the tabular data is successfully extracted. However, it is still not readable.
Hence in the next steps I will be orgainising the data into a structured format using the pandas library. This step involves converting the raw table into a DataFrame, which makes the data more readable and easier to manipulate. 

```python
data_info = table.find_all('th')
data_info = [title.text for title in data_info]
df = pd.DataFrame(columns = data_info)
df
```

Firstly, I have extracted all the table headers. I did this by identifying the table header elements and retrieving their text content.

This prepares the framework for populating the Dataframe with the data from the table rows.

```python
column_data = soup.find_all('tr')
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data

df
```

Now I have extracted the data from each row of the table and populated the DataFrame. I did this by identifying the td tags and adding the data as new rows in the DatFrame.

Now the tabular data we extracted is systematically organised and easy to understand.

#### 3) Checking for out of bound values and validation

By checking for out of bound values I can check if data aligns with expected range with the medical standards. Out-of-bound values could indicate errors during data collection, entry, or preprocessing, which can lead to misleading results or biased analysis. By identifying and addressing these values, I can improve data quality, maintain the integrity of your analysis, and ensure that conclusions drawn are reliable and accurate. 

```python
#Converting platelets from raw counts to kiloplatelets/mL
df1['platelets'] = df1['platelets'] / 1000

#Defining valid ranges for numeric columns
valid_ranges = {
    'age': (40, 95),  # Age range
    'creatinine_phosphokinase': (23, 7861),  # CPK enzyme levels
    'ejection_fraction': (14, 80),  # Ejection fraction percentage range
    'platelets': (25.01, 850.00),  # Platelets in kiloplatelets/mL
    'serum_creatinine': (0.50, 9.40),  # Creatinine level in mg/dL
    'serum_sodium': (114, 148),  # Sodium levels in mEq/L
    'time': (4, 285)  # Follow-up period in days
}

#Checking for out-of-bound values for each column
out_of_bounds = {}
for column, (lower, upper) in valid_ranges.items():
    # Identify out-of-bound values
    out_of_bounds[column] = df1[(df1[column] < lower) | (df1[column] > upper)]

#Displaying rows with out-of-bound values
for column, rows in out_of_bounds.items():
    if not rows.empty:
        print(f"Out-of-bound values detected in '{column}':")
        print(rows)
    else:
        print(f"No out-of-bound values detected in '{column}'.")
```

The results indicate that out-of-bound values were detected in the 'serum_sodium' column. To address this, I will clip these values to the defined lower and upper bounds, ensuring that the data remains within the given range while preserving as much of the dataset as possible for analysis.

Clipping out-of-bound values was chosen over removing rows to preserve the integrity and completeness of the dataset. By clipping, I can ensure that extreme values are brought within acceptable ranges without discarding potentially valuable information. This approach allows the analysis to remain robust while adhering to statistically valid ranges for the variables.

```python
#To clip the out-of-bound values
for column, (lower, upper) in valid_ranges.items():
    df1[column] = df1[column].clip(lower=lower, upper=upper)
```

Now I will move on to check more the out-of-bound values for the boolean features

```python
#Define valid values for binary and Boolean columns
valid_values = {
    'anaemia': [0, 1],
    'high_blood_pressure': [0, 1],
    'diabetes': [0, 1],
    'sex': [0, 1],
    'smoking': [0, 1],
    'DEATH_EVENT': [0, 1]
}

#Check for invalid values
invalid_values = {}
for column, valid_vals in valid_values.items():
    invalid_values[column] = df1[~df1[column].isin(valid_vals)]

#Display rows with invalid values
for column, rows in invalid_values.items():
    if not rows.empty:
        print(f"Invalid values detected in '{column}':")
        print(rows)

#Summarize validity checks
for column, valid_vals in valid_values.items():
    print(f"Checking '{column}':")
    print(f"Unique values: {df1[column].unique()}")
    print(f"Values outside {valid_vals}: {(~df1[column].isin(valid_vals)).sum()}")  # Corrected invalid count
```

This step validates binary and Boolean columns and ensures data integrity by confirming that only the expected values, such as 0 and 1, are present. This step is essential to detect errors in data collection or processing, maintain consistency for statistical analysis or modeling, and ensure the results remain interpretable. 

In this case, the checks revealed no invalid values, indicating that the binary columns are clean and reliable, which allows for accurate and meaningful analysis without concerns of inconsistencies or data errors.
